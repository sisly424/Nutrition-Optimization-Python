***Explanation of Algebraic Formulation Elements***

$$
\begin{array}{rll}
x_i & = \text{number of servings of food item } i \text{ consumed per day} \\
c_i & = \text{cost of food item } i \text{ per serving} \\
N_{ni} & = \text{amount of nutrient } n \text{ in one serving of food item } i \\
\text{Min}_n & = \text{minimum daily requirement for nutrient } n \\
\text{Max}_n & = \text{maximum daily intake allowed for nutrient } n \\
N_{\text{calories}, i} & = \text{amount of calories in one serving of food item } i \\
T_{\text{calories}} & = \text{total daily calorie intake from all food items} \\
N_{\text{protein}, i} & = \text{amount of protein in one serving of food item } i \\
T_{\text{protein}} & = \text{total daily protein intake from all food items} \\
\end{array}
$$

***Objective***

The objective is to minimize the total cost of the food, subject to the following constraints:

- Calories intake between 1800 and 2400 kcal.
- Fat intake between 60 and 95 grams.
- Sodium intake between 1200 and 2200 mg.
- Carbohydrates, fiber, protein, vitamins, and minerals within the required range.
- No single food item should make up more than 30% of the total calories or protein intake
- Non negative.

```python
import pandas as pd
import gurobipy as gp
from gurobipy import GRB
```

```python
file_path = 'food_data.xlsx'
xls = pd.ExcelFile(file_path)

sheet1 = pd.read_excel(xls, 'Sheet1')  # Nutrient limits
sheet2 = pd.read_excel(xls, 'Sheet2')  # Food items and nutritional info

print("Sheet1 Columns:", sheet1.columns.tolist())
print("Sheet2 Columns:", sheet2.columns.tolist())
```

```python
nutrient_limits = {}
for index, row in sheet1.iterrows():
    nutrient_limits[row['Nutrient']] = (row['Min'], row['Max'])
```

```python
foods = sheet2['Food']
calories = sheet2['Calories (kcal)']
fat = sheet2['Fat (g)']
sodium = sheet2['Sodium (mg)']
carbs = sheet2['Carbs (g)']
fiber = sheet2['Fiber (g)']
protein = sheet2['Protein (g)']
vitamin_a = sheet2['Vitamin A (IU)']
vitamin_c = sheet2['Vitamin C (mg)']
calcium = sheet2['Calcium (mg)']
iron = sheet2['Iron (mg)']
cost = sheet2['Cost ($/serving)']
```

```python
model = gp.Model('diet')
servings = model.addVars(len(foods), vtype=GRB.CONTINUOUS, name="servings")
model.setObjective(gp.quicksum(servings[i] * cost[i] for i in range(len(foods))), GRB.MINIMIZE)
```

**Nutrient constraints**

```python
model.addConstr(gp.quicksum(servings[i] * calories[i] for i in range(len(foods))) >= nutrient_limits['Calories (kcal)'][0], "cal_min")
model.addConstr(gp.quicksum(servings[i] * calories[i] for i in range(len(foods))) <= nutrient_limits['Calories (kcal)'][1], "cal_max")
model.addConstr(gp.quicksum(servings[i] * fat[i] for i in range(len(foods))) >= nutrient_limits['Fat (g)'][0], "fat_min")
model.addConstr(gp.quicksum(servings[i] * fat[i] for i in range(len(foods))) <= nutrient_limits['Fat (g)'][1], "fat_max")
model.addConstr(gp.quicksum(servings[i] * sodium[i] for i in range(len(foods))) >= nutrient_limits['Sodium (mg)'][0], "sodium_min")
model.addConstr(gp.quicksum(servings[i] * sodium[i] for i in range(len(foods))) <= nutrient_limits['Sodium (mg)'][1], "sodium_max")
model.addConstr(gp.quicksum(servings[i] * carbs[i] for i in range(len(foods))) >= nutrient_limits['Carbs (g)'][0], "carbs_min")
model.addConstr(gp.quicksum(servings[i] * carbs[i] for i in range(len(foods))) <= nutrient_limits['Carbs (g)'][1], "carbs_max")
model.addConstr(gp.quicksum(servings[i] * fiber[i] for i in range(len(foods))) >= nutrient_limits['Fiber (g)'][0], "fiber_min")
model.addConstr(gp.quicksum(servings[i] * fiber[i] for i in range(len(foods))) <= nutrient_limits['Fiber (g)'][1], "fiber_max")
model.addConstr(gp.quicksum(servings[i] * protein[i] for i in range(len(foods))) >= nutrient_limits['Protein (g)'][0], "protein_min")
model.addConstr(gp.quicksum(servings[i] * protein[i] for i in range(len(foods))) <= nutrient_limits['Protein (g)'][1], "protein_max")
model.addConstr(gp.quicksum(servings[i] * vitamin_a[i] for i in range(len(foods))) >= nutrient_limits['Vitamin A (IU)'][0], "vitamin_a_min")
model.addConstr(gp.quicksum(servings[i] * vitamin_a[i] for i in range(len(foods))) <= nutrient_limits['Vitamin A (IU)'][1], "vitamin_a_max")
model.addConstr(gp.quicksum(servings[i] * vitamin_c[i] for i in range(len(foods))) >= nutrient_limits['Vitamin C (mg)'][0], "vitamin_c_min")
model.addConstr(gp.quicksum(servings[i] * vitamin_c[i] for i in range(len(foods))) <= nutrient_limits['Vitamin C (mg)'][1], "vitamin_c_max")
model.addConstr(gp.quicksum(servings[i] * calcium[i] for i in range(len(foods))) >= nutrient_limits['Calcium (mg)'][0], "calcium_min")
model.addConstr(gp.quicksum(servings[i] * calcium[i] for i in range(len(foods))) <= nutrient_limits['Calcium (mg)'][1], "calcium_max")
model.addConstr(gp.quicksum(servings[i] * iron[i] for i in range(len(foods))) >= nutrient_limits['Iron (mg)'][0], "iron_min")
model.addConstr(gp.quicksum(servings[i] * iron[i] for i in range(len(foods))) <= nutrient_limits['Iron (mg)'][1], "iron_max")
```

**No single food should contribute more than 30% of total calories and protein**

```python
total_calories = gp.quicksum(servings[i] * calories[i] for i in range(len(foods)))
total_protein = gp.quicksum(servings[i] * protein[i] for i in range(len(foods)))

for i in range(len(foods)):
    model.addConstr(servings[i] * calories[i] <= 0.30 * total_calories, f"cal_limit_{i}")
    model.addConstr(servings[i] * protein[i] <= 0.30 * total_protein, f"protein_limit_{i}")
```

**Print the results**

```python
# Optimize the model
model.optimize()

if model.status == GRB.OPTIMAL:
    for i in range(len(foods)):
        if servings[i].x > 0:
            print(f"{foods[i]}: {servings[i].x:.2f} servings")
    print(f"Total Cost: ${model.objVal:.2f}")
else:
    print("No optimal solution found.")
```

```python
import pandas as pd
import gurobipy as gp
from gurobipy import GRB

# Load the Excel file data (replace with your file path if necessary)
file_path = 'food_data.xlsx'
xls = pd.ExcelFile(file_path)

# Read the necessary sheets
sheet1 = pd.read_excel(xls, 'Sheet1')  # Nutrient limits
sheet2 = pd.read_excel(xls, 'Sheet2')  # Food items and nutritional info

nutrient_limits = {}
for index, row in sheet1.iterrows():
    nutrient_limits[row['Nutrient']] = (row['Min'], row['Max'])

food_categories = {
    "Broccoli": "Vegetable",
    "Carrots, Raw": "Vegetable",
    "Corn": "Vegetable",
    "Lettuce, Iceberg,Raw": "Vegetable",
    "Peppers, Sweet, Raw": "Vegetable",
    "Tomato,Red,Ripe,Raw": "Vegetable",
    "Tofu": "Protein",
    "Roasted Chicken": "Protein",
    "Poached Eggs": "Protein",
    "Scrambled Eggs": "Protein",
    "Turkey": "Protein",
    "Beef": "Protein",
    "Pork": "Protein",
    "White Tuna in Water": "Protein",
    "Apple, Raw, w/Skin": "Fruit",
    "Banana": "Fruit",
    "Grapes": "Fruit",
    "Kiwifruit, Raw, Fresh": "Fruit",
    "Oranges": "Fruit",
    "Potatoes, Baked": "Grain/Pasta",
    "Bagels": "Grain/Pasta",
    "Wheat Bread": "Grain/Pasta",
    "White Bread": "Grain/Pasta",
    "Spaghetti W/ Sauce": "Grain/Pasta",
    "Oatmeal": "Grain/Pasta",
    "Couscous": "Grain/Pasta",
    "White Rice": "Grain/Pasta",
    "Macaroni, cooked": "Grain/Pasta",
    "2% Lowfat Milk": "Dairy",
    "Skim Milk": "Dairy"
}

# Apply the mapping to the 'Food' column to create the 'type' variable
sheet2['Type'] = sheet2['Food'].map(food_categories)

foods = sheet2['Food']
calories = sheet2['Calories (kcal)']
fat = sheet2['Fat (g)']
sodium = sheet2['Sodium (mg)']
carbs = sheet2['Carbs (g)']
fiber = sheet2['Fiber (g)']
protein = sheet2['Protein (g)']
vitamin_a = sheet2['Vitamin A (IU)']
vitamin_c = sheet2['Vitamin C (mg)']
calcium = sheet2['Calcium (mg)']
iron = sheet2['Iron (mg)']
cost = sheet2['Cost ($/serving)']
types = sheet2['Type']  

# Initialize the Gurobi model
model = gp.Model('diet')

days = 7

# Decision variables: servings of each food item for each day
servings = model.addVars(days, len(foods), vtype=GRB.CONTINUOUS, name="servings")

# Binary variables to indicate if a food item is selected on a given day
selected = model.addVars(days, len(foods), vtype=GRB.BINARY, name="selected")

# Objective: Minimize total cost over all days
model.setObjective(gp.quicksum(servings[d, i] * cost[i] for d in range(days) for i in range(len(foods))), GRB.MINIMIZE)

# Nutrient constraints for each day
for d in range(days):
    model.addConstr(gp.quicksum(servings[d, i] * calories[i] for i in range(len(foods))) >= nutrient_limits['Calories (kcal)'][0], f"cal_min_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * calories[i] for i in range(len(foods))) <= nutrient_limits['Calories (kcal)'][1], f"cal_max_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * fat[i] for i in range(len(foods))) >= nutrient_limits['Fat (g)'][0], f"fat_min_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * fat[i] for i in range(len(foods))) <= nutrient_limits['Fat (g)'][1], f"fat_max_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * sodium[i] for i in range(len(foods))) >= nutrient_limits['Sodium (mg)'][0], f"sodium_min_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * sodium[i] for i in range(len(foods))) <= nutrient_limits['Sodium (mg)'][1], f"sodium_max_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * carbs[i] for i in range(len(foods))) >= nutrient_limits['Carbs (g)'][0], f"carbs_min_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * carbs[i] for i in range(len(foods))) <= nutrient_limits['Carbs (g)'][1], f"carbs_max_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * fiber[i] for i in range(len(foods))) >= nutrient_limits['Fiber (g)'][0], f"fiber_min_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * fiber[i] for i in range(len(foods))) <= nutrient_limits['Fiber (g)'][1], f"fiber_max_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * protein[i] for i in range(len(foods))) >= nutrient_limits['Protein (g)'][0], f"protein_min_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * protein[i] for i in range(len(foods))) <= nutrient_limits['Protein (g)'][1], f"protein_max_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * vitamin_a[i] for i in range(len(foods))) >= nutrient_limits['Vitamin A (IU)'][0], f"vitamin_a_min_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * vitamin_a[i] for i in range(len(foods))) <= nutrient_limits['Vitamin A (IU)'][1], f"vitamin_a_max_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * vitamin_c[i] for i in range(len(foods))) >= nutrient_limits['Vitamin C (mg)'][0], f"vitamin_c_min_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * vitamin_c[i] for i in range(len(foods))) <= nutrient_limits['Vitamin C (mg)'][1], f"vitamin_c_max_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * calcium[i] for i in range(len(foods))) >= nutrient_limits['Calcium (mg)'][0], f"calcium_min_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * calcium[i] for i in range(len(foods))) <= nutrient_limits['Calcium (mg)'][1], f"calcium_max_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * iron[i] for i in range(len(foods))) >= nutrient_limits['Iron (mg)'][0], f"iron_min_day_{d}")
    model.addConstr(gp.quicksum(servings[d, i] * iron[i] for i in range(len(foods))) <= nutrient_limits['Iron (mg)'][1], f"iron_max_day_{d}")

# Link servings and selected variables
for d in range(days):
    for i in range(len(foods)):
        model.addConstr(servings[d, i] <= selected[d, i] * 1000, f"link_servings_selected_day_{d}_food_{i}")
        model.addConstr(servings[d, i] >= selected[d, i] * 0.1, f"link_servings_selected_day_{d}_food_{i}_min")

# Ensure at least one food item is different each day
for d in range(1, days):
    model.addConstr(gp.quicksum(selected[d, i] - selected[d-1, i] for i in range(len(foods))) >= 1, f"different_food_day_{d}")

# Cumulative constraint to ensure diversity over the week
for i in range(len(foods)):
    if types[i] != "Dairy":
        model.addConstr(gp.quicksum(selected[d, i] for d in range(days)) <= 4, f"cumulative_food_{i}")

# Type diversity constraint
food_types = sheet2['Type'].unique()  # Get unique types

for d in range(days):
    for food_type in food_types:
        type_indices = [i for i in range(len(foods)) if types[i] == food_type]
        model.addConstr(gp.quicksum(servings[d, i] for i in type_indices) >= 1, f"type_diversity_{food_type}_day_{d}")

# Optimize the model
model.optimize()

if model.status == GRB.OPTIMAL:
    print("Optimal meal plan:")
    total_cost = 0
    for d in range(days):
        day_cost = sum(servings[d, i].x * cost[i] for i in range(len(foods)))
        total_cost += day_cost
        print(f"Day {d + 1} (Cost: ${day_cost:.2f}):")
        for i in range(len(foods)):
            if servings[d, i].x > 0:
                print(f"  {foods[i]}: {servings[d, i].x:.2f} servings")
    print(f"Total Cost: ${total_cost:.2f}")
else:
    print("No optimal solution found.")
```
